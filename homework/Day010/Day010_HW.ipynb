## requests + BeautifulSoup


```python
import requests
from  bs4 import  BeautifulSoup

rp = requests.get("https://www.cupoy.com/home")
rp_b = BeautifulSoup(rp.content)
print(type(rp_b),rp_b.find('title').text)
```

## Grab + PyQuery

```python
from  grab import Grab 
from  pyquery import PyQuery as pq

g = Grab()
grp = g.go("http://www.cupoy.com/home")
print(grp.body)
p = pq(grp.body)
print(type(p('title')) , p('title').text())

```
