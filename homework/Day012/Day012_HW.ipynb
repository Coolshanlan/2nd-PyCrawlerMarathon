### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup 
datap = {'offset':50,
'tPage': 3,
'tFile':'20200309.xml',
'tOt': 0,
'tSi': 1000,
'tAr': 0}
# r = requests.post("https://www.ettoday.net/show_roll.php",data=datap)
# rb =BeautifulSoup(r.text)
# print([d.find_all('em')[0].text for d in rb.find_all('h3')])
r = requests.get("https://www.ettoday.net/news/news-list.htm")
rb =BeautifulSoup(r.text)
print([d.text.replace(' ','@').replace('\u3000',' ').split('@') for d in rb.find(class_="part_list_2").findChildren("h3")])
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
post_list =[{'date':d.find('span').text,'tag':d.find('em').text,'content':d.find('a').text.replace('\u3000',' ')} for d in rb.find(class_="part_list_2").findChildren("h3")]
print(post_list)
```

### ③ 哪一個類別的文章最多

```python
tags_count={}
tags_list = list(set([t["tag"] for t in post_list]))
for pl in post_list :
    t=pl["tag"]
    tags_count[tags_list.index(t)] = tags_count.get(tags_list.index(t),0)+1
print(tags_count)
# 轉dict_items
# for tck,tcv in tags_count.items():
#     print(f"{tags_list[tck]}:{tcv}")
# 不轉型
for tc in tags_count:
    print(f"{tags_list[tc]}:{tags_count.get(tc,-1)}")
    
# index sort
# index_sort = sorted(range(len(tags_count.items())-1),key=lambda d:-list(tags_count.values())[d])
# print(index_sort)
# print(tags_list[list(tags_count.keys())[index_sort[0]]],list(tags_count.items())[index_sort[0]][1])

# 直接對別的tags_list sort
tags_count_l=sorted(tags_list,key=lambda d:-tags_count.get(tags_list.index(d),-1))
print(tags_count_l)
print(f"{tags_count_l[0]}:{tags_count[tags_list.index(tags_count_l[0])]}")

# sort by value in dict
tags_count=sorted(tags_count.items(),key=lambda d:-d[1])
print(tags_count)
print(f"{tags_list[tags_count[0][0]]}:{tags_count[0][1]}")

```
