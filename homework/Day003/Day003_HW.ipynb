### 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異


```python
from urllib.request import urlretrieve
import zipfile 
import os
file_url =  "http://opendata.cwb.gov.tw/govdownload?dataid=F-D0047-093&authorizationkey=rdec-key-123-45678-011121314"
os.makedirs( './Data', exist_ok=True )
data_file = urlretrieve(file_url,'./Data/example.zip')
with zipfile.ZipFile('./Data/example.zip') as zip_file:
    zip_file.extractall('./Data')
    zip_file.close()
```

```python
with open('./Data/TAIWAN_72hr_CH.xml','r',encoding='utf-8-sig') as fos:
    fs = fos.read()
    print(fs)
```

```python
import xmltodict

with open('./Data/TAIWAN_72hr_CH.xml','r',encoding='utf-8-sig') as fos:
    fxml = dict(xmltodict.parse(fos.read()))
    print(fxml['cwbopendata']['dataset']['locations']['location'][3]['locationName'][:2] == "新竹")
```

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？
# 找尋有高雄地區的檔案
import os
files = os.listdir('./Data')
findlist=[]
for file_name in files:
    with open(f'./Data/{file_name}',encoding="utf-8-sig") as fos:
        fxml = xmltodict.parse(fos.read())
        if 'locationsName' in list(fxml['cwbopendata']['dataset']['locations'].keys()):
        # print(fxml['cwbopendata']['dataset']['locations'].keys())
            if fxml['cwbopendata']['dataset']['locations']['locationsName'][:2] == "高雄":
                findlist.append(file_name)
print(findlist)
```

```python
#列出數量 與名稱
with open(f'./data/{findlist[0]}',encoding="utf-8-sig") as fos:
    fxml = dict(xmltodict.parse(fos.read()))
    print(len(fxml['cwbopendata']['dataset']['locations']['location']))
    print([d['locationName'] for d in fxml['cwbopendata']['dataset']['locations']['location']])
```

```python
with open(f'./data/{findlist[0]}',encoding="utf-8-sig") as fos:
    fxml = dict(xmltodict.parse(fos.read()))
    print(len(fxml['cwbopendata']['dataset']['locations']['location']))
    print([d['weatherElement'][0] for d in fxml['cwbopendata']['dataset']['locations']['location']])
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度
with open(f'./data/{findlist[0]}',encoding="utf-8-sig") as fos:
    fxml = dict(xmltodict.parse(fos.read()))
    print([[d['locationName'],d['weatherElement'][0]['time'][0]['dataTime'],d['weatherElement'][0]['time'][0]['elementValue']['value']] for d in fxml['cwbopendata']['dataset']['locations']['location']])
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

with open(f'./data/{findlist[0]}',encoding="utf-8-sig") as fos:
    fxml = dict(xmltodict.parse(fos.read()))
    print([[d['dataTime'],d['elementValue']['value']] for d in fxml['cwbopendata']['dataset']['locations']['location'][0]['weatherElement'][0]['time']])
```
